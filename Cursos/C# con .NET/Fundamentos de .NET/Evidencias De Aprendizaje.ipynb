## Compilación y ejecución de código
1. Digitamos el comando `dotnet run` para ejecutar el proyecto, esto nos sirve para comprobar que el proyecto se ejecuta correctamente.

![](img/exec%20program.png)

2. Digitamos el comando `dotnet build` para compilar el proyecto, esto nos sirve para verificar que toda la sintaxis del proyecto esta correcta y no tenga problemas con dependencias.

![](img/Compilacion%20app.png)
