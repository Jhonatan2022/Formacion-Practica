# Extensiones usadas en Visual Studio Code
## Configuración de python
* [Python](https://marketplace.visualstudio.com/items?itemName=ms-python.python)

1. Nos dirigimos a la seción de `View` y seleccionamos `Command Palette` o presionamos `Ctrl + Shift + P` y escribimos `Python: Select Interpreter` y seleccionamos la versión de python que queremos usar.

![](img/Select%20Interpreter.png)



## Configuración de Java
* [Java Extension Pack](https://marketplace.visualstudio.com/items?itemName=vscjava.vscode-java-pack)
* [JDK](https://www.oracle.com/co/java/technologies/downloads/#jdk20-windows)

1. Instalamos el JDK y nos dirigimos a las variables de entorno y agregamos a path la ruta de nuestro JDK que se encuentra en la carpeta de `C:\Program Files\Java\jdk-20\bin`

![](img/Jdk%20Enviroment.png)

2. Cerramos nuestro editor de codigo y ya podremos ejecutar código de java.

![](img/Java%20test.png)



## Configuración de PHP
* [PHP Server](https://marketplace.visualstudio.com/items?itemName=brapifra.phpserver)
* [XAMPP](https://www.apachefriends.org/es/index.html)

1. Nos dirigimos a `Settings` y buscamos `PHP Server` y seleccionamos `PHP Path` y buscamos la ruta de nuestro `php.exe` que se encuentra en la carpeta de `xampp\php\php.exe`

- + En el apartado `PHP config path` debe ir: `C:\xampp\php\php.ini`
- + En el apartado `PHP Path` debe ir: `C:\xampp\php\php.exe`
