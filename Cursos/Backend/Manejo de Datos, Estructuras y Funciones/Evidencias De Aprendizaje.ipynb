```js
// Le pedimos la edad al usuario
let edad = prompt("¿Cuál es tu edad?");

// Si es menor de 18 años, no puede pasar
if (edad < 18) {
  console.log("No puedes pasar al bar");
} else {
  console.log("Puedes pasar al bar");
}

// Representamos el anterior ejercicio pero en un switch
let edad = prompt("¿Cuál es tu edad?");

switch (edad) {
  case "18":
    console.log("Puedes pasar al bar");
    break;
  case "17":
    console.log("No puedes pasar al bar");
    break;
  default:
    console.log("No puedes pasar al bar");
    break;
}

for (var i = 0; i <= 5; i++) {
  console.log("imprime el for de 0 a " + "= " + i);
}


// While
let i = 0;
while (i <= 5) {
  console.log("imprime el while de 0 a " + "= " + i);
  i++;
}
```


```py
# Creamos una funcion privada
# Solo se puede acceder a ella dentro de este archivo
# _ significa que es privada
def _privada():
    print("Soy una funcion privada")


# Creamos una funcion publica
# Se puede acceder a ella desde cualquier archivo
# No tiene ningun caracter especial
def publica():
    print("Soy una funcion publica")
```
