## Formulas y su estructura

```js
=SUMA(C3:C7) // Suma
=SUMA(C3:C7)-SUMA(D3:D7) // Suma
=D7/C7 // División
=F7*C7 // Multiplicación
```	
![](img/Suma.png)



## Formulas aritméticas básicas

```js
=CONTARA(B3:B7) // Contar celdas no vacias
=CONTAR(D3:D7) // Contar celdas no vacias
=PROMEDIO(D3:H3) // Promedio
=REDONDEAR(I3;0) // Redondear
=REDONDEAR.MENOS(I4;0) // Redondear
=SUMAPRODUCTO(C3:C7;F3:F7)/5 // Suma producto
```
![](img/Redondeo.png)



## Subtotales
1. Nos dirigimos a `Datos` y seleccionamos `Subtotales`

```js
=SUBTOTALES(9;C3:C153) // La función SUBTOTALES nos permite realizar operaciones de subtotal en una lista o base de datos que tiene filas agrupadas
```
![](img/subtotal.png)



## Formula de fechas

```js
=DIA(A3) // Dia
=MES(A3) // Mes
=AÑO(A3) // Año
=DIASEM(A3,2) // Dia de la semana
=FECHA(G3;F3;E3) // Fecha
=HOY() // Fecha actual
```
![](img/Fechas.png)



## Formulas de texto

```js
=NOMPROPIO(A3) // Nombre propio de la celda (Nos sirve para corregir nombres)
=ESPACIOS(A3) // Elimina espacios en blanco de la celda
=EXTRAE(A3;1;1) // Extrae un caracter de la celda (1: Inicio, 1: Cantidad de caracteres)
=TEXTO(A3;"00") // Formatea el texto de la celda (00: Formato de salida)
=CONCAT(C3;"-";E3;"-";H3) // Concatena texto de celdas
// Otra forma de hacerlo
=C7&"-"&E7&"-"&H7 // Concatena texto de celdas
```
![](img/Busqueda%20de%20texto.png)



## Formula aleatoria

```js
=ALEATORIO.ENTRE(1;100) // Genera un numero aleatorio entre 1 y 100
```


## Formulas Lógicas

```js
=A3=B3 // Compara si A3 es igual a B3
=Y(A3=B3;A3=C3) // Compara si A3 es igual a B3 y si A3 es igual a C3
=O(A3=B3;A3=C3) // Compara si A3 es igual a B3 o si A3 es igual a C3
```
